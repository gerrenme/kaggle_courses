```python
def remean_points(row):
    row.points = row.points - review_points
    return row

wr.apply(remean_points, axis="columns")
```

```python
def count_tf(frame):
    res_t, res_f = 0, 0
    for index, row in frame.iterrows():
        res_t += row["description"].count("tropical")
        res_f += row["description"].count("fruity")
    
    return pd.Series(data=[res_t, res_f])

descriptor_counts = count_tf(wr)
```
