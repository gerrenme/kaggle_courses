```python
flight_data = pd.read_csv("datasets/flight_delays.csv")

flight_data
```

```python
plt.figure(figsize=(14, 6))

plt.title("Average arrival delay for Spirit Airlines flights by month")
plt.ylabel("Arrival delay (per minutes)")
plt.xlabel("Month")

colors = ['grey' if (x <= flight_data["NK"].mean()) else 'red' for x in flight_data["NK"]]

sns.barplot(x=flight_data.index, y=flight_data["NK"], palette=colors)
```

```python
plt.figure(figsize=(14, 6))

plt.title("Average arrival delay for Each Airline by month")
plt.xlabel("Airline")
plt.ylabel("Month")

sns.heatmap(data=flight_data, annot=True)
```
