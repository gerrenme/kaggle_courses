```python
import pandas as pd
pd.plotting.register_matplotlib_converters()

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
fifa_data = pd.read_csv("datasets/fifa.csv", index_col="Date", parse_dates=True)
```

```python
fifa_data.head(10)
```

```python
plt.figure(figsize=(16, 6))

sns.lineplot(data=fifa_data)
```

```python
spotify_data = pd.read_csv("datasets/spotify.csv", index_col="Date", parse_dates=True)
```

```python
spotify_data.head(5)
```

```python
spotify_data.tail(5)
```

```python
plt.figure(figsize=(16, 8))
plt.title("Daily Global Streams of Popular Songs in 2017-2018")
plt.xlabel("Date (year-moth)")
plt.ylabel("Listenings in mln")

sns.lineplot(data=spotify_data)
```

```python
plt.figure(figsize=(12, 6))
plt.title("Daily Global Streams of 'Shape of You' and 'Dispacito' in 2017-2018")

plt.xlabel("Date (year-month)")
plt.ylabel("Listenings in mln")

sns.lineplot(data=spotify_data["Shape of You"], label="Shape of You")
sns.lineplot(data=spotify_data["Despacito"], label="Despacito")
```

```python
flight_data = pd.read_csv("datasets/flight_delays.csv")

flight_data
```

```python
plt.figure(figsize=(14, 6))

plt.title("Average arrival delay for Spirit Airlines flights by month")
plt.ylabel("Arrival delay (per minutes)")
plt.xlabel("Month")

colors = ['grey' if (x <= flight_data["NK"].mean()) else 'red' for x in flight_data["NK"]]

sns.barplot(x=flight_data.index, y=flight_data["NK"], palette=colors)
```

```python
plt.figure(figsize=(14, 6))

plt.title("Average arrival delay for Each Airline by month")
plt.xlabel("Airline")
plt.ylabel("Month")

sns.heatmap(data=flight_data, annot=True)
```

```python
insurance_data = pd.read_csv("datasets/insurance.csv")

insurance_data.head(10)
```
