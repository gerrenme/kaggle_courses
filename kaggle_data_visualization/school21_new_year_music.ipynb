```python
new_year = pd.read_csv("datasets/new_year_music_feedback.csv", index_col="ID", parse_dates=True)
```

```python
new_year.head(15)
```

```python
new_year.rename(columns={" ": "answer_creation_time", 
                         "        ": "is_music", 
                         "   / ? / ": "morning_music", 
                         "   / ? / ": "evening_music", 
                         "      -       /  .    YouTube": "music_wishes", 
                         "         (c 25    ) /  18 ": "date_start18", 
                         "         (c 25    ) /  19 ": "date_start19", 
                         "         (c 25    ) /  20 ": "date_start20", 
                         "         (c 25    ) /  21 ": "date_start21", 
                         "         (c 25    ) /  22 ": "date_start22", 
                         "         (c 25    ) /  23 ": "date_start23", 
                         "         (c 25    ) /  24 ": "date_start24", 
                         "        ? /  1 ": "date_end1", 
                         "        ? /  2 ": "date_end2", 
                         "        ? /  3 ": "date_end3", 
                         "        ? /  4 ": "date_end4", 
                         "        ? /  5 ": "date_end5", 
                         "        ? /  6 ": "date_end6", 
                         "        ? /  7 ": "date_end7", 
                         " ,         ? /  7   12 ": "time_start7", 
                         " ,         ? /  8   12 ": "time_start8", 
                         " ,         ? /  9   12 ": "time_start9", 
                         " ,         ? /  10   12 ": "time_start10", 
                         " ,         ? /  17   22 ": "time_end17", 
                         " ,         ? /  18   22 ": "time_end18", 
                         " ,         ? /  19   22 ": "time_end19", 
                         " ,         ? /  20   22 ": "time_end20", 
                         " ,         ? /  21   22 ": "time_end21"}, inplace=True)
```

```python
new_year.columns
```

```python
new_year["is_music_on"] = np.where(new_year["is_music"] == "  ", True, False)
new_year.drop(columns=["is_music"], inplace=True)

new_year.rename(columns={"is_music_on": "is_music"}, inplace=True)
new_year['is_music'] = new_year['is_music'].astype('category')

new_year['answer_creation_time'] = pd.to_datetime(new_year['answer_creation_time'])

new_year["answer_date"] = new_year['answer_creation_time'].dt.date
new_year["answer_time"] = new_year['answer_creation_time'].dt.time
```

```python
new_year.head(1)
```

```python
plt.figure(figsize=(14, 6))
total_voices = new_year["is_music"].value_counts()
total_true = total_voices[True]
total_false = total_voices[False]
tt_percent = round(total_true / (total_true + total_false) * 100, 2)

plt.title(f"  {total_voices.sum()} : {total_true} ({tt_percent}%)   {total_false} ({round(100 - tt_percent, 2)}%)  ")
plt.xlabel("(True), (False)")
plt.ylabel(" ")

sns.set_style("whitegrid")
sns.countplot(x="is_music", data=new_year, hue="is_music", palette=["red", "blue"])

plt.savefig("figures/is_music.png")
```

```python
new_year.columns
```

```python
from collections import defaultdict

dates = defaultdict(int)

for col in new_year.columns:
    if col[:-2] == "date_start":
        dates["dec " + col[-2:]] += new_year[col].notna().sum()

dates_to_start = pd.DataFrame(data=list(dates.items()), columns=["date_start", "amount"])

plt.figure(figsize=(14, 6))
plt.title("      ")
plt.xlabel(",    ")
plt.ylabel("    ")
plt.xticks(rotation=45)

sns.set_style("whitegrid")
custom_palette = ['#FF1BB4', '#700040', '#310070', '#004770', '#007013', '#347000', '#703100']
sns.barplot(data=dates_to_start, x="date_start", y="amount", hue="date_start", palette=custom_palette)

plt.savefig("figures/date_start.png")
```

```python
dates = defaultdict(int)

for col in new_year.columns:
    if col[:-1] == "date_end":
        dates["jan " + col[-1:]] += new_year[col].notna().sum()

dates_to_start = pd.DataFrame(data=list(dates.items()), columns=["date_start", "amount"])

plt.figure(figsize=(14, 6))
plt.title("     ")
plt.xlabel(",   ")
plt.ylabel("    ")
plt.xticks(rotation=45)

sns.set_style("whitegrid")
custom_palette = ['#703100', '#700040', '#310070', '#004770', '#007013', '#347000', '#FF1BB4']
sns.barplot(data=dates_to_start, x="date_start", y="amount", hue="date_start", palette=custom_palette)

plt.savefig("figures/date_end.png")
```

```python
hours = defaultdict(int)

for col in new_year.columns:
    if "time_start" in col:
        hours["from " + col[col.find("time_start") + len("time_start")] + " to 12"] += new_year[col].notna().sum()

dates_to_start = pd.DataFrame(data=list(hours.items()), columns=["hour_start", "amount"])

plt.figure(figsize=(14, 6))
plt.title("      ( - )")
plt.xlabel(",    ")
plt.ylabel("    ")
plt.xticks(rotation=45)

sns.set_style("whitegrid")
custom_palette = ['#FF1BB4', '#700040', '#310070', '#004770', '#007013', '#347000', '#703100']
sns.barplot(data=dates_to_start, x="hour_start", y="amount", hue="hour_start", palette=custom_palette)

plt.savefig("figures/time_start.png")
```

```python
hours = defaultdict(int)

for col in new_year.columns:
    if "time_end" == col[:-2]:
        hours[f"from {col[-2:]} to 22"] += new_year[col].notna().sum()

dates_to_start = pd.DataFrame(data=list(hours.items()), columns=["hour_end", "amount"])

plt.figure(figsize=(14, 6))
plt.title("      ( - )")
plt.xlabel(",    ")
plt.ylabel("    ")
plt.xticks(rotation=45)

sns.set_style("whitegrid")
custom_palette = ['#FF1BB4', '#700040', '#310070', '#004770', '#007013', '#347000', '#703100']
sns.barplot(data=dates_to_start, x="hour_end", y="amount", hue="hour_end", palette=custom_palette)

plt.savefig("figures/time_end.png")
```
